# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a popular Python web framework that allows developers to build web applications quickly and easily. It is a lightweight, flexible, and scalable micro web framework that is used to build REST APIs, web services, and web applications.

##### Some advantages of using Flask are:

__Lightweight__: Flask is a micro web framework, which means it has minimal dependencies and is easy to learn and use.

__Flexible__: Flask is highly customizable and allows developers to choose the tools and libraries they want to use in their projects.

__Scalable__: Flask can handle small to large scale applications with ease.

__Built-in Development Server__: Flask comes with a built-in development server, which makes it easy to test and debug web applications.

__Python__: Flask is written in Python and integrates well with other Python libraries and frameworks, making it a natural choice for Python developers.

__Active Community__: Flask has a large and active community of developers who contribute to its development and provide support to others through forums and online resources.

# Q3. What is App routing in Flask? Why do we use app routes?

In Flask, App routing refers to the process of defining URL patterns (or routes) that the application can respond to. When a user makes a request to a specific URL, the Flask application will determine which function should be called to handle the request based on the URL pattern.

App routing is a fundamental part of Flask and is used to map URLs to specific functions in the application. It is accomplished using the @app.route() decorator, which is used to specify the URL pattern that the decorated function should respond to.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint and can also take arguments to generate URLs dynamically.

```python
#code

from flask import Flask
from flask import url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/user/<name>')
def user_profile(name):
    return f'This is the profile page for {name}.'

if __name__ == '__main__':
    with app.test_request_context():
        # generate a URL for the home page
        home_url = url_for('home')
        print(home_url)

        # generate a URL for the user_profile page with a username
        user_profile_url = url_for('user_profile', name)
        print(user_profile_url) 
        app.run(host="0.0.0.0")
```
